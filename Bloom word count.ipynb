```python
import pandas as pd
import re
import spacy
import nltk
from nltk.stem import SnowballStemmer
```

```python
df = pd.read_json('corpus,json')
```

```python
df['text_processed'] = df[0].apply(lambda x: re.sub(r'[,.!?]', '', x))
df['text_processed'] = df['text_processed'].apply(lambda x: x.lower())
```

```python
nlp = spacy.load("en_core_web_sm")
stemmer = SnowballStemmer("english")

def stemming_text(text):
    words = nltk.word_tokenize(text)
    stemmed_text = " ".join([stemmer.stem(word) for word in words])
    return stemmed_text


df['stemmed_text'] = df['text_processed'].apply(stemming_text)
```

```python
dic = {'remember':['recognizing','recalling'],
'understand':['interpreting','exemplifying','classifying','summarizing','inferring','comparing','explaining'],
'apply':['executing','implementing'],'analyze':['differentiating','organizing','attributing'],'evaluate':['checking'
,'critiquing'],'create':['generating','planning','producing']}


for key in dic:
    stemmed_lst = [stemmer.stem(word) for word in dic[key]]
    dic[key] = stemmed_lst

def word_count(text,word_lst):
    
    count = 0
    for word in word_lst:
        count += text.count(word)
        
    return count

for key in dic:
    df[key] = df['text_processed'].apply(lambda x: word_count(x,dic[key]))
    
```
