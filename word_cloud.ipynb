```python
import pandas as pd
import re
import nltk
from wordcloud import WordCloud
```

```python
df = pd.read_json('corpus,json')
```

```python
df
```

```python
df['text_processed'] = df[0].map(lambda x: re.sub('[,\.!?\n]', '', x))
df['text_processed'] = df['text_processed'] .map(lambda x: re.sub('learning goal', '', x))
df['text_processed'] = df['text_processed'] .map(lambda x: x.lower())
```

```python
long_string = ','.join(list(df['text_processed'].values))
wordcloud = WordCloud(background_color="white", max_words=5000, contour_width=3, contour_color='steelblue')
wordcloud.generate(long_string)
wordcloud.to_image()
```
